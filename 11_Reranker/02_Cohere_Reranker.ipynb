````python
def pretty_print_docs(docs):
    print(
        f"\n{'-' * 100}\n".join(
            [f"Document {i+1}:\n\n" + d.page_content for i, d in enumerate(docs)]
        )
    )
````

````python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_cohere import CohereEmbeddings

# ë¬¸ì„œ ë¡œë“œ
# Google Drive ë§ˆìš´íŠ¸ í›„ ì‹¤ì œ íŒŒì¼ ê²½ë¡œë¥¼ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤.
# ì˜ˆì‹œ: /content/drive/MyDrive/Your_Folder/appendix-keywords.txt
# TODO: ì‚¬ìš©ìëŠ” ì´ ê²½ë¡œë¥¼ ì‹¤ì œ Google Drive ë‚´ íŒŒì¼ ê²½ë¡œë¡œ ìˆ˜ì •
file_path = "/content/drive/MyDrive/Colab Notebooks/Lang_chain/11_Reranker/data/appendix-keywords.txt"
documents = TextLoader(file_path).load()

# í…ìŠ¤íŠ¸ ë¶„í• ê¸° ì„¤ì •
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)

# ë¬¸ì„œ ë¶„í• 
texts = text_splitter.split_documents(documents)

# ê²€ìƒ‰ê¸° ì´ˆê¸°í™”
retriever = FAISS.from_documents(
    texts, CohereEmbeddings(model="embed-multilingual-v3.0",
                            cohere_api_key=cohere_api_key)
).as_retriever(search_kwargs={"k": 10})

print("ğŸ‰ Cohere ì„ë² ë”© ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ ë¦¬íŠ¸ë¦¬ë²„ê°€ ì„±ê³µì ìœ¼ë¡œ ì´ˆê¸°í™”ë˜ì—ˆìŠµë‹ˆë‹¤!")
````
